```python
import pandas as pd
import matplotlib.pyplot as plt

# Sentiment Analysis
from textblob import TextBlob
```

```python
# -------------------------------
# STEP 2 : Interview Questions
# -------------------------------

questions = [

    "What is Machine Learning?",

    "Explain Supervised Learning.",

    "What is Python used for?",

    "Difference between AI and ML?",

    "What is NLP?"
]

# Store scores
scores = []

# -------------------------------
# STEP 3 : Start Interview
# -------------------------------

print("================================")
print("WELCOME TO AI INTERVIEW BOT")
print("================================")

candidate_name = input("\nEnter Your Name: ")

print(f"\nHello {candidate_name}!")
print("Interview Started...\n")

# -------------------------------
# STEP 4 : Ask Questions
# -------------------------------

for i, question in enumerate(questions):

    print(f"\nQuestion {i+1}:")
    print(question)

    # User answer
    answer = input("\nYour Answer: ")

    # -------------------------------
    # STEP 5 : NLP Analysis
    # -------------------------------

    # TextBlob sentiment analysis
    analysis = TextBlob(answer)

    polarity = analysis.sentiment.polarity

    # -------------------------------
    # STEP 6 : Generate Score
    # -------------------------------

    # Positive detailed answer -> better score

    if len(answer.split()) > 5:

        if polarity >= 0:
            score = 10
        else:
            score = 6

    else:
        score = 4

    scores.append(score)

    print(f"\nAI Score: {score}/10")

# -------------------------------
# STEP 7 : Final Result
# -------------------------------

total_score = sum(scores)

average_score = total_score / len(scores)

print("\n================================")
print("INTERVIEW COMPLETED")
print("================================")

print(f"\nCandidate Name: {candidate_name}")

print(f"Total Score: {total_score}/50")

print(f"Average Score: {round(average_score,2)}")

# -------------------------------
# STEP 8 : Performance Feedback
# -------------------------------

if average_score >= 8:
    print("\nFeedback: Excellent Performance!")

elif average_score >= 6:
    print("\nFeedback: Good Performance!")

else:
    print("\nFeedback: Needs Improvement!")

# =====================================================
# DATA VISUALIZATION
# =====================================================

# -------------------------------
# GRAPH 1 : Question Wise Score
# -------------------------------

plt.figure(figsize=(8,5))

question_numbers = [
    "Q1",
    "Q2",
    "Q3",
    "Q4",
    "Q5"
]

plt.bar(
    question_numbers,
    scores
)

plt.title("Question Wise Performance")

plt.xlabel("Questions")

plt.ylabel("Scores")

plt.ylim(0,10)

plt.show()

# -------------------------------
# GRAPH 2 : Pie Chart
# -------------------------------

labels = ['Obtained Score', 'Remaining']

values = [
    total_score,
    50 - total_score
]

plt.figure(figsize=(6,6))

plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%'
)

plt.title("Interview Performance")

plt.show()
```
